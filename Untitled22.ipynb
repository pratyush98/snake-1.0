```python
def unit_vector(vector):
    return vector / np.linalg.norm(vector)

def angle_between(v1, v2):
    v1_u = unit_vector(v1)
    v2_u = unit_vector(v2)
    if v2[0]==1 and v2[1]==0:
        if v1[1]>0:
            return np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))
        else:
            return -1*np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))
    if v2[0]==-1 and v2[1]==0:
        if v1[1]<0:
            return np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))
        else:
            return -1*np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))
    if v2[0]==0 and v2[1]==1:
        if v1[0]>0:
            return np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))
        else:
            return -1*np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))
    if v2[0]==0 and v2[1]==-1:
        if v1[0]<0:
            return np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))
        else:
            return -1*np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))
```

```python
orient1=[]
orient2=[]
orient3=[]
orient4=[]
ctr=0
for i,j,m,n in zip(df1['foodx'],df1['foody'],df1['snake_head_x_'],df1['snake_head_y_']):
    veca=[(i[0]-m[0]),(j[0]-n[0])]
    vecb=[df1['velocityx_'][ctr][0],df1['velocityy_'][ctr][0]]
    orient1.append(angle_between(veca,vecb))
    veca=[(i[1]-m[0]),(j[1]-n[0])]
    orient2.append(angle_between(veca,vecb))
    veca=[(i[2]-m[0]),(j[2]-n[0])]
    orient3.append(angle_between(veca,vecb))
    veca=[(i[3]-m[0]),(j[3]-n[0])]
    orient4.append(angle_between(veca,vecb))
    ctr += 1
```

```python
df1['orient1']=orient1
df1['orient2']=orient2
df1['orient3']=orient3
df1['orient4']=orient4
```
